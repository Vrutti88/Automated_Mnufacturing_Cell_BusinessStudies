This notebook analyzes the financial, technical and business impact of implementing an Automated Manufacturing Cell in TechParts Manufacturing Ltd.

## 2. Problem Description

### Background

TechParts Manufacturing Ltd. produces precision metal components for the automobile industry.

Currently production is done manually with separate workstations for cutting, drilling and assembly.

This leads to:
- Higher cycle time
- Human errors
- Material handling delays
- Production defects
- Delivery delays

The company spends ₹8 Crores annually on labor, machine idle time and rework.

The company plans to invest ₹4 Crores in an Automated Manufacturing Cell consisting of CNC machines, robotic arms, conveyors and a central control system.

The objective is to improve productivity, reduce costs and increase manufacturing efficiency.

## 3. Business and Financial Analysis

### Question 1 – Annual Savings Calculation

Current labor and rework cost = ₹8 Crores

Cost reduction due to automation = 30%

Savings = 8 × 0.30  
Savings = ₹2.4 Crores

Productivity increase = 25%

This means the company can produce 25% more components using the same resources and time. Increased productivity improves production capacity, reduces delays, and increases potential revenue.

Therefore, the direct annual cost savings from automation is **₹2.4 Crores**, while productivity improvement provides additional indirect benefits.

### Impact of Productivity Increase

Automation also increases productivity by 25%.

This means the company can produce 25% more components using the same time and resources.

Higher productivity leads to:
- Increased production output
- Faster delivery
- Higher potential revenue
- Better utilization of machines

Although productivity increase does not directly reduce costs, it significantly improves overall profitability and operational efficiency.

### Question 2 - Net Benefit Calculation

Formula:
Net Benefit = Total Savings – Investment

Total Savings = ₹2.4 Crores
Investment = ₹4 Crores

Net Benefit = 2.4 – 4
Net Benefit = -1.6 Crores

But note:
Automation also increases productivity by 25%, which will generate extra revenue.

Although the first year shows a negative net benefit of ₹1.6 Crores, the investment will recover in approximately 2 years due to continuous annual savings and increased production output.

## 4. Technical Components

### Key Components of an Automated Manufacturing Cell

1. CNC Machines
Computer Numerical Control machines perform precise cutting and drilling operations automatically.

2. Robotic Arms
Robots handle loading, unloading and assembly tasks which reduces human errors.

3. Conveyor Systems
Conveyors transport materials automatically between machines.

4. Sensors
Sensors monitor machine status, temperature and product quality.

5. Central Control System
A computer system that coordinates all machines and operations.

6. Manufacturing Execution System (MES)
Tracks production data and machine performance in real time.

7. Safety Systems
Emergency stop systems and safety barriers ensure worker safety.

## 5. Python Data Simulation

```python
import pandas as pd
import numpy as np

# number of machines
n = 50

data = {
    "Machine_ID": np.arange(1, n+1),
    "Cycle_Time_Min": np.random.uniform(5, 15, n),
    "Defect_Rate_%": np.random.uniform(1, 5, n),
    "Production_Volume": np.random.randint(100, 500, n),
    "Downtime_Hours": np.random.uniform(1, 10, n),
    "Operational_Cost": np.random.uniform(5000, 20000, n)
}

df = pd.DataFrame(data)

df.head()
```

Synthetic production data is generated using Python libraries such as NumPy and Pandas.

NumPy generates random numerical values such as cycle time, defect rate and downtime.

Pandas organizes the generated data into a structured dataset (DataFrame) for analysis.

This allows simulation of real manufacturing data without accessing actual company records.

```python
import matplotlib.pyplot as plt

plt.hist(df["Cycle_Time_Min"])
plt.title("Machine Cycle Time Distribution")
plt.xlabel("Cycle Time")
plt.ylabel("Frequency")
plt.show()
```
